```python
# Lookinng at the counts it will be difficult to get 200 for all levels
```

```python
def get_ratings_reviews(df):
  level_reviews = []
  for i in df['reviews_rating'].unique():
    # shuffling the dataset
    df_1 = df.sample(frac=1)
    df_1 = df_1[df_1['reviews_rating'] == i]
    if len(df_1) > 200:
      reviews = df_1['reviews_text'][:200].tolist()
      level_reviews.append(reviews)
    else:
      reviews = df_1['reviews_text'].tolist()
      level_reviews.append(reviews)
  
  return level_reviews

level_reviews = get_ratings_reviews(df)
len(level_reviews)
```

```python
df_1 = df[df['reviews_rating'] == 1]
rev1 = df_1['reviews_text'].tolist()
```

```python
from bertopic import BERTopic
```

```python
topic_model = BERTopic(verbose=True, embedding_model='all-mpnet-base-v2', min_topic_size=30)

topics, _ = topic_model.fit_transform(rev1)
```
